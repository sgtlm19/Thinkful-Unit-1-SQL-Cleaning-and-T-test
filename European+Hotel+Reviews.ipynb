# Intermediate Visualization Drill with Seaborn#

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import European Hotel Review Dataset from Kaggle ##

```python
hotels = pd.read_csv('Hotel_Reviews.csv')
```

```python
hotels.info()
```

```python
hotels.head(n=30)
```

### Choose one variable and plot that variable four different ways. ###

```python
#Chose to plot the Review Total Negative Word Counts
negative_words = hotels['Review_Total_Negative_Word_Counts']
```

```python
plt.hist(negative_words, color='blue', alpha=0.5, label='Negative Word Count Distribution')
plt.xlabel('Number of Negative Words')
plt.legend(loc='upper right')
plt.title('Negative Word Count in European Hotel Reviews')
plt.show()
```

```python
sns.distplot(negative_words)
```

```python
sns.boxplot(negative_words)
```

```python
sns.kdeplot(negative_words, shade=True)
```

### Choose two continuous variables, and plot them three different ways. ###

```python
g = sns.lmplot(y='Review_Total_Negative_Word_Counts', # Variable 1.
               x='Review_Total_Positive_Word_Counts', # Variable 2.
               data=hotels, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Negative Word Count")
g.set_xlabels("Positive Word Count")
plt.title('Scatterplot: Negative Word Count by Positive Word Count')
plt.show()
```

```python
g = sns.lmplot(y='Review_Total_Negative_Word_Counts', # Variable 1.
               x='Review_Total_Positive_Word_Counts', # Variable 2.
               data=hotels, # Data
               fit_reg=True, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Negative Word Count")
g.set_xlabels("Positive Word Count")
plt.title('Scatterplot: Negative Word Count by Positive Word Count w/Regression')
plt.show()
```

```python
sns.jointplot(x="Review_Total_Positive_Word_Counts", y="Review_Total_Negative_Word_Counts", data=hotels, kind="kde")
```

```python
positive_words = hotels['Review_Total_Positive_Word_Counts']
```

```python
plt.hist(negative_words, color='red',  alpha=.5,label='Negative Words')
plt.hist(positive_words, color='blue', alpha=.5, label='Positive Words')
plt.xlabel('Word Count')
plt.legend(loc='upper right')
plt.title('Histogram of Positive and Negative Word Counts')
plt.show()
```

### Choose one continuous variable and one categorical variable, and plot them six different ways. ###

```python
Ireland_UAE = hotels[(hotels['Reviewer_Nationality'] == ' Ireland ') | (hotels['Reviewer_Nationality'] == ' United Arab Emirates ')]
```

```python
ax = sns.boxplot(x='Reviewer_Nationality',y='Review_Total_Negative_Word_Counts',hue='Reviewer_Nationality',data=Ireland_UAE)  
plt.title('Boxplot by Nationality')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Negative Word Count')

plt.show()
```

```python
sns.set(style="darkgrid")

g = sns.factorplot(x='Reviewer_Nationality', y= 'Review_Total_Negative_Word_Counts', hue='Reviewer_Nationality', data=Ireland_UAE,
                   size=6, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Negative Word Counts")
g.set_xlabels("")
plt.title('Barplot: Negative Word Counts Ireland vs UAE')
plt.show()
```

```python
# Setting the overall aesthetic.
sns.set(style="whitegrid")

g = sns.factorplot(x='Reviewer_Nationality', y='Review_Total_Negative_Word_Counts', hue='Reviewer_Nationality', data=Ireland_UAE,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Total Negative Word Counts")
g.set_xlabels("")
plt.title('Pointplot: Total Negative Word Counts Ireland vs UAE')
plt.show()
```

```python
Ireland_Word_Counts = Ireland_UAE.loc[(Ireland_UAE['Reviewer_Nationality']==' Ireland '), 'Review_Total_Negative_Word_Counts']
UAE_Word_Counts = Ireland_UAE.loc[(Ireland_UAE['Reviewer_Nationality']== ' United Arab Emirates '), 'Review_Total_Negative_Word_Counts']
```

```python
plt.hist(Ireland_Word_Counts, color='red',  alpha=.5,label='Ireland')  # alpha just controls the opacity
plt.hist(UAE_Word_Counts, color='blue', alpha=.5,label='UAE')
plt.xlabel('Country')
plt.legend(loc='upper right')
plt.title('Negative Word Counts Ireland vs. UAE')
plt.show()
```

```python
sns.set(style="ticks")

sns.boxplot(x='Review_Total_Negative_Word_Counts', y="Reviewer_Nationality", data=Ireland_UAE,
            whis=np.inf, palette="vlag")
plt.title('Horizontal Box Plot Negative Word Count Ireland vs. UAE')

```

```python
sns.distplot(Ireland_Word_Counts, hist=False, color="g", kde_kws={"shade": True})
sns.distplot(UAE_Word_Counts, hist=False, color="b", kde_kws={"shade": True})
plt.title('Kernel Density Estimate Plot Ireland vs. UAE')
```
